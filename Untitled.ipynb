### Convertition de la données ZEOP en BAN

```python
def Convert_ZEOP_to_FormatBAN(Ban_Adress, Zeop_Adress):
    Ban_Adress_Empty = Ban_Adress.iloc[0:0]

    Zeop_Adress_reodered = Ban_Adress_Empty
    Zeop_Adress_reodered['lon'] = Zeop_Adress['x']
    Zeop_Adress_reodered['lat'] = Zeop_Adress['y']
    Zeop_Adress_reodered['nom_voie'] = Zeop_Adress['type_voie']+ " " + Zeop_Adress['nom_voie'].str.lower()
    Zeop_Adress_reodered['numero'] = Zeop_Adress['num_voie']
    Zeop_Adress_reodered['rep'] = Zeop_Adress['cp_no_voie']
    Zeop_Adress_reodered['id'] = Zeop_Adress['imb_id']

    return Zeop_Adress_reodered
```

##### Utilisation de l'algo des distances euclidiennes

```python
import pandas as pd
import numpy as np
```

```python
# Fonction pour calculer les distances
# La df2 sera la donnée où chaque identifiant sera comparé à celui de la BAN (df1)
def find_nearest(df1, df2):
    nearest_points = []
    for i, row2 in df2.iterrows():
        distances = np.sqrt((df1['lon'] - row2['lon'])**2 + (df1['lat'] - row2['lat'])**2)
        nearest_idx = distances.idxmin()  # Index du point le plus proche
        nearest_points.append({'df2_id': row2['id'], 'nearest_df1_id': df1.loc[nearest_idx, 'id'], 
                               'distance': distances[nearest_idx]})
    return pd.DataFrame(nearest_points)

```

Mise à jour de la donnée FIRME en fonction des distances avec la donnée BAN

```python
def ChangeFirmeAdress_With_NeareastCoord( Ban_Adr, Zeop_Adr):
    # Trouver les points les plus proches
    result = find_nearest(Ban_Adr, Zeop_Adr)
    
    Files_Empty = Ban_Adr.iloc[0:0]
    copy_Zeop_Adr = Zeop_Adr.copy(deep=True)
    sum_updated = 0
    # Boucle sur les résultats
    for _, row in result.iterrows():
        if row["distance"] < 0.000050:
            id = row["nearest_df1_id"]
            
            # Extraire l'adresse correspondante dans Ban_Adr
            adr = Ban_Adr[Ban_Adr["id"] == id][["numero", "rep","nom_voie"]]
            #print(adr)
            if not adr.empty:  # Vérifier que les données existent
                adr_num = adr.iloc[0]["numero"]
                adr_rep = adr.iloc[0]["rep"]
                adr_nomVoie = adr.iloc[0]["nom_voie"]
                #print(adr_nomVoie)
                # Mettre à jour Zeop_Adr
                id_zeop = row["df2_id"]
                #print(copy_Zeop_Adr[copy_Zeop_Adr["id"] == id_zeop]["nom_voie"])
                copy_Zeop_Adr.loc[copy_Zeop_Adr["id"] == id_zeop, "numero"] = adr_num
                copy_Zeop_Adr.loc[copy_Zeop_Adr["id"] == id_zeop, "rep"] = adr_rep
                copy_Zeop_Adr.loc[copy_Zeop_Adr["id"] == id_zeop, "nom_voie"] = adr_nomVoie
                #print(copy_Zeop_Adr[copy_Zeop_Adr["id"] == id_zeop]["nom_voie"])
                sum_updated+=1
                #print(copy_Zeop_Adr.loc[copy_Zeop_Adr["id"] == id_zeop])
                Files_Empty = pd.concat([Files_Empty, copy_Zeop_Adr.loc[copy_Zeop_Adr["id"] == id_zeop]])
            #print("-------")
    print(sum_updated)
    return copy_Zeop_Adr, Files_Empty
```

##### Convertion de ZEOP au format BAN, vers ZEOP

```python
def Convert_BAN_to_FormatZEOP(Ban_Format, Zeop_Format):
    Zeop_Adress_Empty = Zeop_Format.iloc[0:0]

    ## ON garde uniqument les varialbes du fichier BAN
    ## ATTENTION RETIRER LES TIRETS

    Ban_To_Zeop = Zeop_Adress_Empty
    #print(Zeop_Format)
    Ban_To_Zeop['x'] = Ban_Format['lon']
    Ban_To_Zeop['y'] = Ban_Format['lat']
    Ban_To_Zeop['num_voie'] = Ban_Format['numero']
    Ban_To_Zeop['cp_no_voie'] = Ban_Format['rep']
    Ban_To_Zeop['type_voie'] = Ban_Format['nom_voie'].str.split().str[0] #Récupérer le premier mot
    Ban_To_Zeop['nom_voie'] = Ban_Format['nom_voie'].str.split().str[1:].str.join(' ').str.upper() #Récupére tous sauf le primier mot + UPPER
    
    Ban_To_Zeop['imb_id'] = Ban_Format['id']

    # Fusion des anciennes données : 
    
    

    return Ban_To_Zeop
```
